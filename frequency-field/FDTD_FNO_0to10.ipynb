## 訓練資料的大小

### 250筆

```python
import meep as mp
import numpy as np
import matplotlib.pyplot as plt
from meep import materials
import torch
import os
```

```python
from FDTD import FDTDSimulator, generate_dataset

NUM_SAMPLES = 250
RATIO       = 0

X_all, Y_all, freq_list = generate_dataset(
    n_samples        = NUM_SAMPLES,          
    freq_min         = 1.0,
    freq_max         = 5.0,
    total_meep_time  = 100,
    num_frames       = 100,         
    resolution       = 333,
    seed             = 42,
    ratio            = RATIO,
    save_path_X      = f"dataset_X_{NUM_SAMPLES}.npy",
    save_path_Y      = f"dataset_Y_{NUM_SAMPLES}.npy",
    save_path_freq   = f"dataset_freq_{NUM_SAMPLES}.npy",
    verbose          = True,
)
```

```python
import FNO

EPOCH = 250

CFG = dict( path_X          = f"dataset_X_{NUM_SAMPLES}.npy",
            path_Y          = f"dataset_Y_{NUM_SAMPLES}.npy",
            train_ratio     = 0.7,
            val_ratio       = 0.15,
            n_modes         = (12, 12),
            n_layers        = 4,
            hidden_channels = 24,
            in_channels     = None,
            out_channels    = None,
            batch_size      = 4,
            lr              = 1e-3,
            weight_decay    = 1e-4,
            scheduler_step  = 60,
            scheduler_gamma = 0.5,
            epochs          = EPOCH,
            seed            = 42,
            device          = "cpu",
            save_dir        = f"checkpoints_{NUM_SAMPLES}",
          )

model, history, test_loader = FNO.train(CFG)

FNO.plot_loss(history)

best_path = os.path.join(CFG["save_dir"], "best_model.pt")

model.load_state_dict(torch.load(best_path, map_location=CFG["device"], weights_only=False))

preds, targets = FNO.test(model, test_loader, CFG)

FNO.plot_prediction(preds, targets, sample_idx=0, frame_idx=-1)
```
